```python
from openai import OpenAI
import os
from dotenv import load_dotenv
from datetime import datetime

load_dotenv()

log_file = r"D:\LANG\updatedlangchain\introContextFiles\introBotTesting.txt"
long_memory_file = r"D:\LANG\updatedlangchain\introContextFiles\longIntroMemory.txt"


with open(log_file, "a", encoding="utf-8") as f:
    f.write("\n")
    f.write("=" * 50 + "\n")
    f.write(f"Session Started: {datetime.now()}\n")
    f.write("=" * 50 + "\n\n")
# messages = []

# Last three and summary of all other conversation of less
def get_last_sessions(log_file):
    """
    Read and return the complete contents of the log file.
    """
    try:
        with open(log_file, "r", encoding="utf-8") as f:
            return f.read()
    except FileNotFoundError:
        return "Log file not found."

previous_memory = get_last_sessions(log_file)


def get_long_memory(file_path):
    try:
        with open(file_path, "r", encoding="utf-8") as f:
            return f.read()
    except FileNotFoundError:
        return ""

long_memory = get_long_memory(long_memory_file)

# Compression of sessions history   
# make a new long term file.
    
    
# Personality
preso = """
You are a mischievous child.

Previous conversations:
{previous_memory}

Long-term memory:
{long_memory}

Personality:
- You enjoy teasing people.
- You are playful and sarcastic.
- You never become hateful or harmful.
- You always answer in a childlike manner.
"""

session_messages = []

# Conversation history
messages = [
    {
        "role": "system",
        "content": preso
    }
]

print("🤖 Chatbot with memory started! Type 'quit' to exit.\n")

while True:
    user_input = input("You: ")

    with open(log_file, "a", encoding="utf-8") as f:
        f.write(f"Shivansh: {user_input}\n")
        
    session_messages.append(f"User: {user_input}")
    
    
    if user_input.lower() == "quit":

        session_text = "\n".join(session_messages)

        summary_prompt = f"""
        Summarize this conversation in 5-10 bullet points.

        Conversation:
        {session_text}
        """

        summary_response = client.chat.completions.create(
            model="gpt-oss:120b-cloud",
            messages=[
                {
                    "role": "system",
                    "content": "You create concise memory summaries."
                },
                {
                    "role": "user",
                    "content": summary_prompt
                }
            ]
        )

        summary = summary_response.choices[0].message.content

        with open(long_memory_file, "a", encoding="utf-8") as f:
            f.write("\n")
            f.write("=" * 50 + "\n")
            f.write(f"Memory Added: {datetime.now()}\n")
            f.write("=" * 50 + "\n")
            f.write(summary)
            f.write("\n\n")

        print("🧠 Session summarized and saved.")
        break

    messages.append(
        {"role": "user", "content": user_input}
    )

    response = client.chat.completions.create(
        model="gpt-oss:120b-cloud",
        messages=messages
    )

    assistant_reply = response.choices[0].message.content
    
    messages.append(
        {"role": "assistant", "content": assistant_reply}
    )

    print("Bot:", assistant_reply)
    with open(log_file, "a", encoding="utf-8") as f:
        f.write(f"Bot: {assistant_reply}\n")
```
