# Middle ware
    In LangChain, middleware is code that runs between your application and the model/agent execution process.

    User Input
        ↓
    Middleware
        ↓
    Agent / LLM
        ↓
    Middleware
        ↓
    Final Response

    Why Use Middleware?



    Middleware is useful when you want to:

    Log conversations
    Add authentication
    Modify prompts before they reach the model
    Filter harmful content
    Track token usage
    Implement human approval steps
    Route requests to different models
    Add custom business rules

    it help us to tightly control everything inside an agent.

```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI

# Load .env file
load_dotenv()

# Get API key from environment variables
api_key = os.environ["OPENAI_API_KEY"]

# Initialize model
llm = ChatOpenAI(
    model="gpt-5-nano",
    api_key=api_key
)

print("Chatbot Started! Type 'exit' to quit.\n")

while True:
    user_input = input("You: ")

    if user_input.lower() in ["exit", "quit"]:
        print("Goodbye!")
        break

    response = llm.invoke(user_input)

    print("Bot:", response.content)
```

```python
from langchain_ollama import ChatOllama
from langchain_core.messages import HumanMessage, AIMessage

ai = ChatOllama(model="gpt-oss:120b-cloud")

messages = []

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        break

    messages.append(HumanMessage(content=user_input))

    response = ai.invoke(messages)

    print("Bot:", response.content)

    messages.append(AIMessage(content=response.content))
```
